```python
all_relations = {}
for set in ['train','test','val']:
    print(len(relationship[set]))
    for image in relationship[set]:
        all_relations[image] = relationship[set][image]
imageIds = list(all_relations.keys())
```

```python
imageIds = [int(i) for i in imageIds]
imageIds.sort()
imageIds = [str(i) for i in imageIds]
random.seed(0)
random.shuffle(imageIds)
train_imageIds = imageIds[:2000]
val_imageIds = imageIds[3000:3500]
test_imageIds = imageIds[4000:4500]
```

```python
newDatarelationship = {}
newDatarelationship['train'] = {}
newDatarelationship['val'] = {}
newDatarelationship['test'] = {}
for image in train_imageIds:
    newDatarelationship['train'][image] = all_relations[image]
for image in val_imageIds:
    newDatarelationship['val'][image] = all_relations[image]
for image in test_imageIds:
    newDatarelationship['test'][image] = all_relations[image]
newDatarelationship['rel_categories'] = relationship['rel_categories']
json.dump(newDatarelationship, open('./data/vg/rel.json', 'w'))
```

```python
train_data = json.load(open('./data/vg/oldtrain.json', 'r'))
val_data = json.load(open('./data/vg/oldval.json', 'r'))
test_data = json.load(open('./data/vg/oldtest.json', 'r'))
train_data.keys()
```

```python
train_data['categories'] == test_data['categories'] == val_data['categories']
```

```python
images_data = train_data['images'] + test_data['images'] + val_data['images']
annotations_data = train_data['annotations'] + test_data['annotations'] + val_data['annotations']
```

```python
train_annotations = []
val_annotations = []
test_annotations = []
for annotation in tqdm(annotations_data):
    if str(annotation['image_id']) in train_imageIds:
        train_annotations.append(annotation)
    elif str(annotation['image_id']) in val_imageIds:
        val_annotations.append(annotation)
    elif str(annotation['image_id']) in test_imageIds:
        test_annotations.append(annotation)
len(train_annotations), len(val_annotations), len(test_annotations)
```

```python
anIds = []
for data in annotations_data:
    anIds.append(data['image_id'])
anIds = np.unique(anIds)
```

```python
train_images = []
val_images = []
test_images = []
for image in tqdm(images_data):
    if str(image['id']) in train_imageIds:
        train_images.append(image)
    elif str(image['id']) in val_imageIds:
        val_images.append(image)
    elif str(image['id']) in test_imageIds:
        test_images.append(image)
len(train_images), len(val_images), len(test_images)
```

```python
train_database = {'images': train_images, 'annotations': train_annotations, 'categories': train_data['categories']}
val_database = {'images': val_images, 'annotations': val_annotations, 'categories': val_data['categories']}
test_database = {'images': test_images, 'annotations': test_annotations, 'categories': test_data['categories']}
json.dump(train_database, open('./data/vg/train.json', 'w'))
json.dump(val_database, open('./data/vg/val.json', 'w'))
json.dump(test_database, open('./data/vg/test.json', 'w'))
```

```python
from collections import defaultdict
image2category = defaultdict(list)
for data in subData_annotation:
    image2category[data['image_id']].append(data['category_id'])
```

```python
frequency = np.zeros((150,150))
for image in image2category:
    for i in image2category[image]:
        for j in image2category[image]:
            frequency[i-1][j-1] += 1
```
